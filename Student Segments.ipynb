```python
import numpy as np  # linear algebra
import pandas as pd  # dad preprocessing
import matplotlib.pyplot as plt  # data visualization
import seaborn as sns  # data visualization
from sklearn.cluster import KMeans  # clulstering
```

```python
df = pd.read_excel('response-excel.xlsx')
```

```python
df.columns
```

```python
df_student_segments = df[["מה המגדר שלך?", "מהו גילך בשנים?", "באיזה תחום אתה מחפש עבודה?",
                          "איזה סוג משרה אתה מחפש?", "מאיפה אתה מעדיף לעבוד?",
                          "האם יש לך ניסיון תעסוקתי קודם בתחום בו אתה מחפש עבודה?", "האם יש לך הסמכות בינלאומיות בתחום בו אתה מחפש עבודה?",
                          "מהי ההשכלה שלך? גם אם הינך במהלך התואר אנא סמן את התשובה"]]
```

```python
df_student_segments = df_student_segments.rename(columns={"מה המגדר שלך?": "Gender",
                                                          "מהו גילך בשנים?": "Age",
                                                          "באיזה תחום אתה מחפש עבודה?": "Job_Categories",
                                                          "איזה סוג משרה אתה מחפש?": "Job_Type",
                                                          "מאיפה אתה מעדיף לעבוד?": "Job_Location",
                                                          "האם יש לך ניסיון תעסוקתי קודם בתחום בו אתה מחפש עבודה?": "Expereince",
                                                          "האם יש לך הסמכות בינלאומיות בתחום בו אתה מחפש עבודה?": "Certifications",
                                                          "מהי ההשכלה שלך? גם אם הינך במהלך התואר אנא סמן את התשובה": "Education"})
```

```python
df_student_segments
```

```python
mask = df_student_segments['Age'] == 'מתחת ל- 18'
df_student_segments = df_student_segments.drop(df_student_segments[mask].index)

df_student_segments= df_student_segments.dropna()
```

```python
# columns exploration
data_types = pd.DataFrame(df_student_segments.dtypes)
data_types.columns = ['DataType']
data_types['CountUnique'] = [df_student_segments[col].nunique() for col in df_student_segments.columns]
data_types['UniqueExamples'] = [df_student_segments[col].unique()[0:10] for col in df_student_segments.columns]
data_types
```

```python
# kmeans clustering
columns_to_cluster = ['Age', 'Job_Categories', 'Job_Type', 'Job_Location', 'Expereince', 'Certifications', 'Education']
df_encoded = pd.get_dummies(df_student_segments[columns_to_cluster])
kmeans = KMeans(n_clusters=2, random_state=0).fit(df_encoded)
df_student_segments['Cluster'] = kmeans.labels_
```

```python
# columns exploration
data_types = pd.DataFrame(df_student_segments.dtypes)
data_types.columns = ['DataType']
data_types['CountUnique'] = [df_student_segments[col].nunique() for col in df_student_segments.columns]
data_types['UniqueExamples'] = [df_student_segments[col].unique()[0:10] for col in df_student_segments.columns]
data_types
```

```python
# Create a new DataFrame with the most frequent values for each column in each cluster
most_frequent = pd.DataFrame(columns=df_student_segments.columns)
for cluster in df_student_segments['Cluster'].unique():
    cluster_df = df_student_segments[df_student_segments['Cluster'] == cluster]
    most_frequent.loc[cluster] = cluster_df.mode().iloc[0]

# Transpose the DataFrame to get the clusters as rows
student_segments = most_frequent.T
student_segments.columns = ['Cluster 0', 'Cluster 1']
```
